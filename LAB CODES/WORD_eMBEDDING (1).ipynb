# **`WORD eMBEDDING`**

```python
sentences = [
    "India is famous for its rich culture and traditions.",
    "India is known for its rich history and culture.",
    "Indian culture is rich and full of traditions.",

    "Cricket is a popular game in India.",
    "Cricket is one of the most loved games in the country.",

    "Technology helps people in daily life.",
    "Modern technology makes daily life easier.",

    "Artificial Intelligence is a part of modern technology.",
    "AI technology helps machines learn and make decisions."
]
```

```python
#IMPORTING BASIC LIBRAIES
import numpy as np
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer#TF-BOW
from sklearn.feature_extraction.text import TfidfVectorizer#TF-IDF
from sklearn.metrics.pairwise import cosine_similarity
```

```python
#TF-BOW Embedding
tf_vectorizer=CountVectorizer()
tf_emb=tf_vectorizer.fit_transform(sentences)
```

```python
#similarity matrix using cosine_similarity
tf_sim=cosine_similarity(tf_emb)
```

```python
tf_sim
```

```python
#TF-IDF Embedding
tridf_vectorizer=TfidfVectorizer()
tridf_emb=tridf_vectorizer.fit_transform(sentences)
```

```python
tfidf_sim=cosine_similarity(tridf_emb)
```

```python
tfidf_sim
```

```python
#Recommend the most matching/similar sentences
def doc_sim(tfidf_sim,sentence):
  for i in range(len(sentences)):
    sim=tfidf_sim[i].copy()
    sim[i]=-1
    top_index=np.argmax(sim)
    print(sentences[i],"--->",sentences[top_index])
```

```python
doc_sim(tfidf_sim,sentences)
```
